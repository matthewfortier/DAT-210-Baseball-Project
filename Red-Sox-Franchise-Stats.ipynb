```python
from lxml import html
import time
import requests
import pandas as pd
```

```python
page = requests.get("http://www.hallofstats.com/franchise/bos#franchise")
tree = html.fromstring(page.content)
```

```python
names = tree.xpath('//ol[@id="player-list"]//li[contains(@class, "franchise")]//strong/text()')
warValues = tree.xpath('//ol[@id="player-list"]//li[contains(@class, "franchise")]//span/text()')
```

```python
names = [n.replace('\n', '') for n in names]
warValues = [n.replace('\n', '') for n in warValues]

hallOfFameDictionary = dict(zip(names, warValues))
```

```python
df = pd.DataFrame({'Name': names, 'WAR': warValues})
df
```

```python
df.to_csv("Franchise-Stats.csv", sep=",", index=False)
```
