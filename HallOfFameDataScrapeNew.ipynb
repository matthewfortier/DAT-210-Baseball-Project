```python
from lxml import html
import time
import requests
import pandas as pd
```

```python
page = requests.get("http://www.hallofstats.com/#hof")
tree = html.fromstring(page.content)
```

```python
names = tree.xpath('//ol[@id="player-list"]//li[not(contains(@class, "not-hof"))]//strong/text()')
warValues = tree.xpath('//ol[@id="player-list"]//li[not(contains(@class, "not-hof"))]//span/text()')
```

```python
names = [n.replace('\n', '') for n in names]
warValues = [n.replace('\n', '') for n in warValues]
print(names, warValues)
```

```python
hallOfFameDictionary = dict(zip(names, warValues))
```

```python
#Easier to put the two lists into dataframe instead of dictionary, since the dictionary loses the ordering
df = pd.DataFrame({'Name': names, 'WAR': warValues})
df
```

```python
df.to_csv("Hall_of_Fame2.csv", sep=",", index=False)
```
